```python
import pandas as pd
import matplotlib.pyplot as pyplot

def plot(x, y, title, xLabel, yLabel):
    fig, axis = pyplot.subplots()
    axis.set_title(title, fontsize=30)
    axis.set_ylabel(yLabel)
    axis.set_xlabel(xLabel)
    axis.scatter(x, y)
    axis.margins(x=0, y=0)
    

df = pd.read_csv("haddadPosts.tsv", "\t")
df.columns = ["id", "message", "date", "shares", "angry_reaction"]
df["date"] = pd.to_datetime(df["date"])
df.head()
xAxis = df["date"]
yAxis = df["shares"]
pyplot.rcParams["figure.figsize"] = [30, 15]
plot(xAxis, yAxis, "Popularidade postagens", "Date", "Shares")
```

```python
def createHistogram(values, title, xLabel, yLabel):
    axes = values.plot(kind="bar")
    axes.set_title(title, fontsize=30)
    axes.set_xlabel(xLabel, fontsize=22)
    axes.set_ylabel(yLabel, fontsize=22)
    pyplot.show()
    
groupByMonth = df.groupby([df["date"].dt.year, df["date"].dt.month])["shares"].mean()
createHistogram(groupByMonth, "Numero de compartilhamentos mensais", "Compartilhamentos", "Data")
```
